# **Data Assimilation - Modified Cholesky**

Nino-Ruiz, E. D., Sandu, A., & Deng, X. (2018). An ensemble Kalman filter implementation based on modified Cholesky decomposition for inverse covariance matrix estimation. SIAM Journal on Scientific Computing, 40(2), A867-A886.

https://epubs.siam.org/doi/abs/10.1137/16M1097031

\begin{eqnarray}
\frac{dx_i}{dt} = (x_{i+1}-x_{i-2})x_{i-1} - x_i + F
\end{eqnarray}

When $F=8$, the model exhibit chaotic behavior

````python
import numpy as np
import seaborn as sns
import scipy as sci
from scipy.integrate import odeint
import matplotlib.pyplot as plt
````

````python
def lorenz96(x, t, args=(8,)):
  n = x.size;
  F = args[0];
  return [(x[(i+1)%n]-x[i-2])*x[i-1]-x[i]+F for i in range(0,n)]; #list comprehension
````

````python
def forecast(x, t):
  return odeint(lorenz96, x, t)[-1,:];
````

````python
def create_initial_ensemble(xb, t, N):
  Xb = [];
  for e in range(0,N):
    xbe_pert = xb+0.01*np.random.randn(n); #e-th pertured ensemble member (no consistent with model dynamics)
    xbe = forecast(xbe_pert, t); #e-th ensemble member (consistent with model dynamics)
    Xb.append(xbe);
  return Xb;
````

````python
def forecast_ensemble(Xb, t, N):
  for e,xbe in enumerate(Xb):
    Xb[e,:] = forecast(Xb[e,:], t);
  return Xb;
````

Create initial settings 

1. The initial condition
2. Initial background
3. Initial background ensemble

````python
#Create initial condition

n = 40; #40 variables
x_random = np.random.randn(n); #pertured random state (no consistent with model dynamics)
t = np.linspace(0,10, 50);
x0 = forecast(x_random, t); #initial condition (consistent with model dynamics)

xt = forecast(x0, t); #reference solution
xb = forecast(x0+0.01*np.random.randn(n), t); #background state (about this state we'll build the initial ensemble)

N = 100;
Xb0 = create_initial_ensemble(xb, t, N);
xt0 = forecast(x0, t);
````

# ${\bf X}^a = {\bf X}^b + {\bf P}^b {\bf H}^T \underbrace{\left ( {\bf R}+{\bf H}{\bf P}^b {\bf H}^T \right )^{-1} \underbrace{\left[{\bf Y}^s-{\bf H}{\bf X}^b \right ]}_{{\bf D}}}_{{\bf Z}} = {\bf X}^b + {\bf P}^b {\bf H}^T {\bf Z}$

````python
p = 0.8; #[0, 1] 0: no observations and 1: full observational network
m = round(p*n); #number of observations m<=n
err_obs = 0.01; #standard deviations of errors (a typical value)
R = (err_obs**2)*np.eye(m, m); #covariance matrix of data errors
M = 20; #number of assimilation cycles
t = [0, 0.1]; #16 hours ~ 8 hours (in practice)
N = 20;
````

````python
xt = np.array(xt0).astype('float32');
Xa = np.array(Xb0.copy()).astype('float32')[:N,:];

err_a = [];
err_b = [];

for k in range(M):
  #Forecast step
  xt = forecast(xt, t);
  Xb = forecast_ensemble(Xa, t, N);
  Pb = np.cov(Xb.T);
  xb = np.mean(Xb, axis=0);
  err_b.append(np.linalg.norm(xb-xt));
 
  #Create observation - data for the assimilation process
  H = np.random.permutation(np.arange(n))[:m]; #we observe a different part of the domain
  y = np.random.multivariate_normal(xt[H], R);

  #Assimilation step
  Ys = np.random.multivariate_normal(y, R, N).T; #synthetic observations
  Ds = Ys - Xb[:,H].T; #Synthetic innovations
  Pa = R + Pb[H,:][:,H]; #Pa = R + H @ Pb @ H.T
  Za = np.linalg.solve(Pa, Ds);
  DX = Pb[:,H] @ Za;
  Xa = Xb + DX.T;
  xa = np.mean(Xa, axis=0);
  err_a.append(np.linalg.norm(xa-xt));
````

# **Local ensemble Kalman filter (LEnKF)**

# ${\bf X}_i^a = {\bf X}_i^b + {\bf P}_i^b {\bf H}_i^T \underbrace{\left ( {\bf R}_i+{\bf H}_i{\bf P}_i^b {\bf H}_i^T \right )^{-1} \underbrace{\left[{\bf Y}_i^s-{\bf H}_i{\bf X}_i^b \right ]}_{{\bf D}_i}}_{{\bf Z}_i} = {\bf X}_i^b + {\bf P}_i^b {\bf H}_i^T {\bf Z}_i$.

````python
def local_EnKF(Xbi, Pbi, Ri, Hi, Di):
  Pai = Ri + Pbi[Hi,:][:,Hi]; #Pa = R + H @ Pb @ H.T
  Zai = np.linalg.solve(Pai, Di.T);
  DXi = Pbi[:,Hi] @ Zai;
  Xai = Xbi + DXi.T;
  return Xai
````

````python
def local_analysis_EnKF(Xb, H, R, y, N, i, r):
  #Subdomain decomposition
  si = [(i+j)%n for j in range(-r,r+1)]
  Xbi = Xb[:, si]
  Pbi = np.cov(Xbi.T)

  #Observations
  oi = np.array([s_i for s_i in si if s_i in H]); #Global index
  Hi = np.array([i for i,s_i in enumerate(si) if s_i in H]); #Local indexes
  mi = len(Hi); #Number of local observations

  if mi>0:
    yi = yz[oi]; #We take the local observations from the model state
    Ri = (err_obs**2)*np.eye(mi, mi); #Local error covariance matrix
    Ysi = np.random.multivariate_normal(yi, Ri, N); #Synthetic observations
    Di = Ysi - Xbi[:, Hi]; #Innovation matrix (local)
    Xai = local_EnKF(Xbi, Pbi, Ri, Hi, Di);
  else:
    Xai = Xbi;
  return Xai;
````

# **Local Ensemble Transform Kalman Filter**

https://www2.atmos.umd.edu/~ekalnay/pubs/harlim_hunt05.pdf

Szunyogh, I., Kostelich, E. J., Gyarmati, G., Kalnay, E., Hunt, B. R., Ott, E., ... & Yorke, J. A. (2008). A local ensemble transform Kalman filter data assimilation system for the NCEP global model. Tellus A: Dynamic Meteorology and Oceanography, 60(1), 113-130.

1. ${\bf x} = \overline{\bf x}^b + \boldsymbol{\Delta} {\bf w}$, wherein $\boldsymbol{\Delta} = {\bf X}^b - \overline{\bf x}^b$ (broadcast)
2. $\mathcal{J}({\bf x}) = \mathcal{J}(\overline{\bf x}^b + \boldsymbol{\Delta} {\bf w}) = \frac{N-1}{2} \left\| {\bf w} \right \|^2 + \frac{1}{2} \left\| {\bf d} - {\bf Q}{\bf w}\right \|^2_{{\bf R}^{-1}}$ where ${\bf d} = {\bf y} - {\bf H} \overline{\bf x}^b$
3. ${\bf w}^{*} = \arg \underset{{\bf w}}{\min}\mathcal{J}(\overline{\bf x}^b + \boldsymbol{\Delta} {\bf w})$
4. $\nabla \mathcal{J} = (N-1){\bf w} - {\bf Q}^T {\bf R}^{-1} [{\bf d} - {\bf Q}{\bf w}]$, wherein ${\bf Q} = {\bf H} \boldsymbol{\Delta}$
5. ${\bf w}^{*} = [(N-1){\bf I} + {\bf Q}^T {\bf R}^{-1}{\bf Q} ]^{-1} {\bf Q}^T {\bf R}^{-1} {\bf d}$


6. $\overline{\bf x}^a = \overline{\bf x}^b + \boldsymbol{\Delta} {\bf w}^{*}$

```python
def local_ETKF(Xbi, Ri, Hi, di):
  N,ni = Xbi.shape;
  xmbi = np.mean(Xbi, axis=0); 
  Deli = Xbi - xmbi;
  Qi   = Deli[:, Hi].T; 
  Pai = (N-1)*np.eye(N) + Qi.T @ np.linalg.solve(Ri, Qi); #Pa = R + H @ Pb @ H.T
  zai = np.linalg.solve(Pai, Qi.T @ np.linalg.solve(Ri, di));
  xmai = xmbi.T + Deli.T @ zai
  Wa   = inv_sqrt(Pai); #This internaly multiplies by N-1
  Xai  = xmai.reshape(-1,1) + Deli.T @ Wa;
  return Xai.T;
```

````python
def inv_sqrt(A):
  N = A.shape[0];
  U,S,V = np.linalg.svd(A);
  Pa_sqrt = U@np.diag(np.sqrt((N-1)/S))@U.T
  return Pa_sqrt;
````

````python
def local_ETKF(Xbi, Ri, Hi, di):
  N,ni = Xbi.shape;
  xmbi = np.mean(Xbi, axis=0); 
  Deli = Xbi - xmbi;
  Qi   = Deli[:, Hi].T; 
  Pai = (N-1)*np.eye(N) + Qi.T @ np.linalg.solve(Ri, Qi); #Pa = R + H @ Pb @ H.T
  zai = np.linalg.solve(Pai, Qi.T @ np.linalg.solve(Ri, di));
  xmai = xmbi.T + Deli.T @ zai
  Wa   = inv_sqrt(Pai); #This internaly multiplies by N-1
  Xai  = xmai.reshape(-1,1) + Deli.T @ Wa;
  return Xai.T;
````

````python
def local_analysis_ETKF(Xb, H, R, y, N, i, r):
  #Subdomain decomposition
  si = [(i+j)%n for j in range(-r,r+1)]
  Xbi = Xb[:, si]
  Pbi = np.cov(Xbi.T)

  #Observations
  oi = np.array([s_i for s_i in si if s_i in H]); #Global index
  Hi = np.array([i for i,s_i in enumerate(si) if s_i in H]); #Local indexes
  mi = len(Hi); #Number of local observations

  if mi>0:
    yi = yz[oi]; #We take the local observations from the model state
    Ri = (err_obs**2)*np.eye(mi, mi); #Local error covariance matrix
    xmbi = np.mean(Xbi, axis=0);
    di = yi - xmbi[Hi]; #Innovation matrix (local)
    Xai = local_ETKF(Xbi, Ri, Hi, di); #Xbi, Pbi, Ri, Hi, di
  else:
    Xai = Xbi;
  return Xai;
````

````python
np.random.seed(10)
xt = np.array(xt0).astype('float32');

err_a = [];
err_b = [];

Xa = np.array(Xb0.copy()).astype('float32')[:N,:];

r = 1;

for k in range(0, M):
  
  xt = forecast(xt, t);
  Xb = forecast_ensemble(Xa, t, N);
  #Create observation - data for the assimilation process
  H = np.random.permutation(np.arange(n))[:m]; #we observe a different part of the domain
  H.sort()

  #Synthetic observation to test our method
  y = np.random.multivariate_normal(xt[H], R);
  yz = np.zeros((n,))

  #We project back onto the model space (to make easier the indexation)
  yz[H] = y

  Xa = np.zeros((N,n)); #Local analysis for each model component i
  for i in range(0, n):
    Xai = local_analysis_ETKF(Xb, H, R, y, N, i, r)
    Xa[:,i] = Xai[:,r]; #

  xb = np.mean(Xb, axis=0);
  err_b.append(np.linalg.norm(xb-xt));
  xa = np.mean(Xa, axis=0);
  err_a.append(np.linalg.norm(xa-xt));
````

# **EnKF via a modified Cholesky decomposition**

````python
p = 0.8; #[0, 1] 0: no observations and 1: full observational network
m = round(p*n); #number of observations m<=n
err_obs = 0.01; #standard deviations of errors (a typical value)
R = (err_obs**2)*np.eye(m, m); #covariance matrix of data errors
Rinv = 1/(err_obs**2)*np.ones((m,)); #covariance matrix of data errors
M = 20; #number of assimilation cycles
t = [0, 0.1];
N = 20;
r = 1; #Radius of influence
ind = np.arange(0, n)

````

````python
from sklearn.linear_model import LinearRegression
````

````python
def get_Binv(Db, r):
  N,n = Db.shape
  L = np.zeros((n,n));
  D = np.zeros((n,n)); #This is the inverse of D
  lr = LinearRegression(fit_intercept=False);
  for i in range(0, n):
    si = [(i+j)%n for j in range(-r,r+1)]
    pi = [j for j in si if j<i]
    L[i, i] = 1;
    y = Db[:, i];
    if len(pi)==0: #No hay predecesores
      D[i, i] = 1/np.var(y);
    else:
      X = Db[:, pi]
      lr_fit = lr.fit(X, y);
      L[i, pi] = -lr_fit.coef_;
      erri = y - lr_fit.predict(X);
      D[i, i] = 1/np.var(erri);
  Binv = L.T @ D @ L;
  return Binv;
````

````python
xt = np.array(xt0).astype('float32');
Xa = np.array(Xb0.copy()).astype('float32')[:N,:];

err_a = [];
err_b = [];

r = 3;

for k in range(M):
  #Forecast step
  xt = forecast(xt, t);
  Xb = forecast_ensemble(Xa, t, N);
  xb = np.mean(Xb, axis=0);
  Db = Xb - xb;
  Binv = get_Binv(Db, r);
  err_b.append(np.linalg.norm(xb-xt));
 
  #Create observation - data for the assimilation process
  H = np.random.permutation(np.arange(n))[:m]; #we observe a different part of the domain
  y = np.random.multivariate_normal(xt[H], R);

  #Assimilation step
  Ys = np.random.multivariate_normal(y, R, N).T; #synthetic observations
  Ds = Ys - Xb[:,H].T; #Synthetic innovations
  HRinv = np.zeros((n,));
  HRinv[H] = Rinv;
  HTRinvH = np.diag(HRinv);
  A = Binv + HTRinvH; 
  Q = np.zeros((n, N));
  Q[H, :] = np.linalg.solve(R, Ds);
  Za = np.linalg.solve(A, Q);
  Xa = Xb + Za.T;
  xa = np.mean(Xa, axis=0);
  err_a.append(np.linalg.norm(xa-xt));
````
